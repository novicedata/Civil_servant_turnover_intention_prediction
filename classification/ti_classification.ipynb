```python
import pandas as pd
import numpy as np

'Q_2022.csv'
def preprocessing(file_path):
    print(f'===== 원데이터 불러오기 : {file_path}', '\n')
    raw_data = pd.read_csv(file_path, low_memory = False)
    print(f'원 데이터 : {raw_data.shape}')

    print('===== EM / LM / Z 세대 나누기 =====', '\n')
    raw_data = raw_data[raw_data['dq2'] >= 1981]

    conditions = [
    (raw_data['dq2'] >= 1981) & (raw_data['dq2'] <= 1988),
    (raw_data['dq2'] >= 1989) & (raw_data['dq2'] <= 1995),
    (raw_data['dq2'] >= 1996)
    ]
    
    choices = ['early_M', 'late_M', 'Z']
    
    data = raw_data.copy()
    data.loc[:, 'gen'] = np.select(conditions, choices, default=np.nan)
    print(f'세대 나눈 데이터 : {data.shape}')

    print('===== 불필요 데이터 삭제 1차 =====', '\n')
    remove_col = ['ID', 'q1_1_m1', 'q1_1_m2', 'q5_1_1', 'q5_2_1', 'q5_3_1',
             'q6_1_1', 'q6_2_1', 'q6_3_1', 'q6_4_1', 'A1', 'A2',
             'q9_m1', 'q9_m2', 'q9_m3', 'q13_1_1', 'q13_1_2', 'q13_1_3',
             'q17_m1', 'q17_m2', 'q18_m1', 'q18_m2', 'q28_m1', 'q28_m2', 'q28_m3',
             'B', 'q31_C1', 'q31_C2', 'q38_1', 'q38_2_m1', 'q38_2_m2',
             'dq7', 'DM2', 'DM3', 'DM4', 'DM6', 'DM7', 'WT', 'TWT', 'dq2']

    new_data = data.drop(remove_col, axis=1)
    print(f'세대 + 불필요 칼럼 처리 : {new_data.shape}')

    print('===== 역문항 처리 =====', '\n')
    # 역문항 처리
    columns_to_reverse = ['q8_5', 'q14_4', 'q19_5', 'q33_5', 'q33_6', 'q33_7', 'q33_8']
    
    def reverse(x):
        if x == 5:
            return 1
        elif x == 4:
            return 2
        elif x == 3:
            return 3
        elif x == 2:
            return 4
        else:
            return 5
    
    for column in columns_to_reverse:
        print(f'{column} 처리중...')
        new_data[column] = new_data[column].apply(reverse)

    # 차원 합치기
    print('\n', '===== 유사 문항 평균화 =====', '\n')
    new_data['resource'] = new_data[['q3_1', 'q3_2', 'q3_3']].mean(axis=1)
    new_data['job_autonomy'] = new_data[['q4_1', 'q4_2', 'q4_3']].mean(axis=1)
    new_data['recruitment'] = new_data[['q7_1', 'q7_2', 'q7_3', 'q7_4']].mean(axis=1)
    new_data['promotion'] = new_data[['q8_1', 'q8_2', 'q8_3', 'q8_4', 'q8_5']].mean(axis=1)
    new_data['job_rotation'] = new_data[['q10_1', 'q10_2', 'q10_3']].mean(axis=1)
    new_data['compensation_rewards'] = new_data[['q11_1', 'q11_2', 'q11_3', 'q11_4']].mean(axis=1)
    new_data['welfare'] = new_data[['q13_1', 'q13_2']].mean(axis=1)
    new_data['training'] = new_data[['q14_1', 'q14_2', 'q14_3', 'q14_4']].mean(axis=1)
    new_data['performance_capacity'] = new_data[['q15_1', 'q15_2', 'q15_3']].mean(axis=1)
    new_data['organizational_goals'] = new_data[['q19_1', 'q19_2', 'q19_3', 'q19_4', 'q19_5']].mean(axis=1)
    new_data['leadership'] = new_data[['q20_1', 'q20_2', 'q20_3', 'q20_4', 'q20_5', 'q20_6', 'q20_7']].mean(axis=1)
    new_data['decision_making'] = new_data[['q22_1', 'q22_2', 'q22_3', 'q22_4']].mean(axis=1)
    new_data['change_management'] = new_data[['q23_1', 'q23_2', 'q23_3']].mean(axis=1)
    new_data['collaboration_communication'] = new_data[['q24_1', 'q24_2', 'q24_3']].mean(axis=1)
    new_data['organizational_performance'] = new_data[['q25_1', 'q25_2']].mean(axis=1)
    new_data['performance_management'] = new_data[['q26_1', 'q26_2', 'q26_3', 'q26_4', 'q26_5']].mean(axis=1)
    new_data['job_performance'] = new_data[['q27_1', 'q27_2', 'q27_3']].mean(axis=1)
    new_data['public_service'] = new_data[['q30_1', 'q30_2', 'q30_3', 'q30_4', 'q30_5', 'q30_6']].mean(axis=1)
    new_data['organizational_citizenship_behavior'] = new_data[['q32_1', 'q32_2', 'q32_3', 'q32_4', 'q32_5']].mean(axis=1)
    new_data['innovative_behavior'] = new_data[['q33_1', 'q33_2', 'q33_3', 'q33_4', 'q33_5', 'q33_6', 'q33_7', 'q33_8']].mean(axis=1)
    new_data['public_service_values'] = new_data[['q34_1', 'q34_2', 'q34_3', 'q34_4', 'q34_5', 'q34_6', 'q34_7', 'q34_8']].mean(axis=1)
    new_data['commitment'] = new_data[['q35_1', 'q35_2', 'q35_3', 'q35_4']].mean(axis=1)
    new_data['job_satisfaction'] = new_data[['q36_1', 'q36_2', 'q36_3']].mean(axis=1)
    new_data['public_job_satisfaction'] = new_data[['q37_1', 'q37_2', 'q37_3', 'q37_4']].mean(axis=1)
    new_data['quality_of_life'] = new_data[['q39_1', 'q39_2']].mean(axis=1)

    print('===== 불필요 데이터 삭제 2차 =====', '\n')

    remove_col2 = [ "q3_1", "q3_2", "q3_3",
      "q4_1", "q4_2", "q4_3",
      "q7_1", "q7_2", "q7_3", "q7_4",
      "q8_1", "q8_2", "q8_3", "q8_4", "q8_5",
      "q10_1", "q10_2", "q10_3",
      "q11_1", "q11_2", "q11_3", "q11_4",
      "q13_1", "q13_2",
      "q14_1", "q14_2", "q14_3", "q14_4",
      "q15_1", "q15_2", "q15_3",
      "q19_1", "q19_2", "q19_3", "q19_4", "q19_5",
      "q20_1", "q20_2", "q20_3", "q20_4", "q20_5", "q20_6", "q20_7",
      "q22_1", "q22_2", "q22_3", "q22_4",
      "q23_1", "q23_2", "q23_3",
      "q24_1", "q24_2", "q24_3",
      "q25_1", "q25_2",
      "q26_1", "q26_2", "q26_3", "q26_4", "q26_5",
      "q27_1", "q27_2", "q27_3",
      "q30_1", "q30_2", "q30_3", "q30_4", "q30_5", "q30_6",
      "q32_1", "q32_2", "q32_3", "q32_4", "q32_5",
      "q33_1", "q33_2", "q33_3", "q33_4", "q33_5", "q33_6", "q33_7", "q33_8",
      "q34_1", "q34_2", "q34_3", "q34_4", "q34_5", "q34_6", "q34_7", "q34_8",
      "q35_1", "q35_2", "q35_3", "q35_4",
      "q36_1", "q36_2", "q36_3",
      "q37_1", "q37_2", "q37_3", "q37_4",
      "q39_1", "q39_2"]
    
    final_data = new_data.drop(remove_col2, axis=1)
    print(f'불필요 칼럼 처리2 : {final_data.shape}', '\n')

    print('===== 이직 의도 이진화 =====')
    conditions2 = [
    (final_data['q38'] <= 3),
    (final_data['q38'] > 3)
    ]
    
    choices2 = ['N', 'Y']
    
    final_data2 = final_data.copy()
    final_data2.loc[:, 'y'] = np.select(conditions2, choices2, default=np.nan)
    final_data2 = final_data2.drop(['q38'], axis=1)
    print(f'이직 의도 이진화 후 기존 열 처리 : {final_data2.shape}')
    print(f'결측치 : {final_data2.isnull().sum().sum()}', '\n')
    

    print('..... 전처리 전/후 레이블 길이 비교 .....')
    print(final_data['q38'].head())
    print(final_data2['y'].head())

    Z = final_data2[final_data2['gen'] == 'Z']
    EM = final_data2[final_data2['gen'] == 'early_M']
    LM = final_data2[final_data2['gen'] == 'late_M']
    
    Z_X = Z.drop(['y', 'gen'], axis=1)
    EM_X = EM.drop(['y', 'gen'], axis=1)
    LM_X = LM.drop(['y', 'gen'], axis=1)

    Z_y = Z['y']
    EM_y = EM['y']
    LM_y = LM['y']

    return Z_X, Z_y, LM_X, LM_y, EM_X, EM_y
```

```python
Z_X, Z_y, LM_X, LM_y, EM_X, EM_y = preprocessing('Q_2022.csv')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Z_X, Z_y, test_size = 0.2, random_state = 42, shuffle = False)
```

```python
print(f'X_train : {X_train.shape}, y_train 수 : {len(y_train)}')
print(f'X_test : {X_test.shape}, y_test 수 : {len(y_test)}')
```

```python
y_test_num = [1 if y == 'Y' else 0 for y in y_test]
```

```python
import numpy as np
import pandas as pd
from pycaret.classification import setup, compare_models, predict_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, recall_score, precision_score, f1_score, cohen_kappa_score, matthews_corrcoef

# 평가 지표를 저장할 데이터 구조 초기화
results = {model: [] for model in range(19)}  # 19개 모델을 저장할 딕셔너리

n_iterations = 100
for i in range(n_iterations):
    print(f"Iteration {i+1}/{n_iterations}")

    # PyCaret 설정
    clf1 = setup(data=X_train, target=y_train, verbose=False, html=False, session_id=i)

    # 모든 모델 비교
    best_models = compare_models(sort='AUC', n_select=19, fold=5, cross_validation=True, verbose=False)

    # 각 모델에 대한 예측 및 평가 지표 계산
    for model in best_models:
        predictions = model.predict(X_test)
        actual = y_test_num
        #pred_labels = predictions['prediction_label']
        #pred_scores = predictions['prediction_score']
        accuracy = accuracy_score(actual, predictions)
        auc = roc_auc_score(actual, predictions)
        recall = recall_score(actual, predictions)
        precision = precision_score(actual, predictions)
        f1 = f1_score(actual, predictions)
        kappa = cohen_kappa_score(actual, predictions)
        mcc = matthews_corrcoef(actual, predictions)

        model_name = model.__class__.__name__
        metrics = {
            'Accuracy': accuracy,
            'AUC': auc,
            'Recall': recall,
            'Precision': precision,
            'F1': f1,
            'Kappa': kappa,
            'MCC': mcc
        }

        # 결과 저장
        if model_name not in results:
            results[model_name] = []
        results[model_name].append(metrics)

# 평균 및 표준편차 계산
summary = {}
for model_name, metrics_list in results.items():
    summary[model_name] = {}
    for metric in ['Accuracy', 'AUC', 'Recall', 'Precision', 'F1', 'Kappa', 'MCC']:
        mean_val = np.mean([m[metric] for m in metrics_list])
        std_val = np.std([m[metric] for m in metrics_list])
        summary[model_name][metric] = {'Mean': mean_val, 'Std': std_val}
```

```python
import pandas as pd

# summary 딕셔너리를 데이터프레임으로 변환
data_for_df = []
index_labels = []

for model_name, metrics in summary.items():
    for metric_name, metric_values in metrics.items():
        mean = metric_values['Mean']
        std = metric_values['Std']
        data_for_df.append(f"{mean:.4f} ± {std:.4f}")
        index_labels.append((model_name, metric_name))

# 인덱스 설정 (모델 이름과 메트릭 이름)
multi_index = pd.MultiIndex.from_tuples(index_labels, names=["Model", "Metric"])

# 데이터프레임 생성
df = pd.DataFrame(data_for_df, index=multi_index, columns=["Value"])

# 데이터프레임을 보기 좋게 조정 (unstack 사용)
df_unstacked = df.unstack(level='Metric')

# 지정된 순서
order = ["Accuracy", "AUC", "Recall", "Precision", "F1", "Kappa", "MCC"]

# 지정된 순서에 따라 열을 재정렬
df_unstacked = df_unstacked["Value"][order]
```
